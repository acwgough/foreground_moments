# Power spectra


```python
%pylab inline

import healpy as hp
```

```python
nside_corse = 32
nside_fine = 128
```

```python
Lmax = 384
ells = np.arange(1,Lmax+1)
```

```python
#now define different kinds of power spectra

cls_uniform = np.ones_like(ells)
cls_linear = c_ell = np.arange(1,Lmax+1)
cls_random = ones_like(ells) + rand(Lmax)

def powerlaw(ell, beta):
    return (ell/80.)**beta

beta1 = -3
beta2 = 3
cls_power1 = powerlaw(ells, beta1)
cls_power2 = powerlaw(ells, beta2)
cls_power1[0] = 0
cls_power1[1] = 0

```

```python
fig, ax = plt.subplots(figsize=(15, 10))
subplot(3,2,1)
semilogy(ells, cls_uniform)
title('Uniform')

subplot(3,2,2)
semilogy(ells, cls_linear)
title('Linear')

subplot(3,2,3)
semilogy(ells, cls_power1)
title('Power, $beta <0$')

subplot(3,2,4)
semilogy(ells, cls_power2)
title('Power, $beta >0$')

subplot(3,2,5)
semilogy(ells, cls_random)
title('Random')

tight_layout()
```

```python
#make some corse maps
cmap_uniform = hp.synfast(cls_uniform, nside_corse);
cmap_linear = hp.synfast(cls_linear, nside_corse);
cmap_power1 = hp.synfast(cls_power1, nside_corse);
cmap_power2 = hp.synfast(cls_power2, nside_corse);
cmap_random = hp.synfast(cls_random, nside_corse);

#make some fine maps
fmap_uniform = hp.synfast(cls_uniform, nside_fine);
fmap_linear = hp.synfast(cls_linear, nside_fine);
fmap_power1 = hp.synfast(cls_power1, nside_fine);
fmap_power2 = hp.synfast(cls_power2, nside_fine);
```

```python
hp.mollview(cmap_uniform, title='Uniform')
hp.mollview(cmap_linear, title='Linear')
hp.mollview(cmap_power1, title='Power, beta<0')
hp.mollview(cmap_power2, title='Power, beta>0')
hp.mollview(cmap_random, title='Random')
```
